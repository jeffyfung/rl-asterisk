```python
import matplotlib.pyplot as plt
import numpy as np
import torch
%matplotlib inline

# Plot the reward over episodes
epi_lens = np.load("data/base/epi_lens.npy")
epi_rtns = np.load("data/base/epi_rtns.npy")
epoch_epi_lens = np.load("data/base/epoch_epi_lens.npy")
epoch_rtns = np.load("data/base/epoch_rtns.npy")

epi_lens_td = np.load("data/td_target_adv/epi_lens.npy")
epi_rtns_td = np.load("data/td_target_adv/epi_rtns.npy")
epoch_epi_lens_td = np.load("data/td_target_adv/epoch_epi_lens.npy")
epoch_rtns_td = np.load("data/td_target_adv/epoch_rtns.npy")
```

```python
print(epi_lens.shape)
print(epi_rtns.shape)
print(epoch_epi_lens.shape)
print(epoch_rtns.shape)
```

```python
epi_rtns_ = torch.tensor(epi_rtns.mean(axis=0)).unfold(0, 10, 1).numpy().mean(axis=1)
epi_rtns_td_ = torch.tensor(epi_rtns_td.mean(axis=0)).unfold(0, 10, 1).numpy().mean(axis=1)
# plt.plot(epi_rtns.mean(axis=1))
plt.plot(epi_rtns_, label="GAE lambda")
plt.plot(epi_rtns_td_, label="TD residual")
plt.xlabel("Episode")
plt.ylabel("Average Return")
plt.legend(loc="lower right")
```

```python
epi_lens_ = torch.tensor(epi_lens.mean(axis=0)).unfold(0, 10, 1).numpy().mean(axis=1)
epi_lens_td_ = torch.tensor(epi_lens_td.mean(axis=0)).unfold(0, 10, 1).numpy().mean(axis=1)

plt.plot(epi_lens_, label="GAE lambda")
plt.plot(epi_lens_td_, label="TD residual")
plt.xlabel("Episode")
plt.ylabel("Episode Length")
plt.legend(loc="lower right")
```

```python
plt.plot(epoch_rtns.mean(axis=0))
plt.xlabel("Epochs")
plt.ylabel("Rewards")
```
